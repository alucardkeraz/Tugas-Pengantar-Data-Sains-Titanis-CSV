```python
import pandas as pd
import numpy as np



df = pd.read_csv('C:/Users/M Rizky Saputra/Downloads/datatni.csv')

print("=== Data Calon TNI ===")
print(df)
print("")
print("")


print("=== Statistik Dasar Tinggi Badan ===")
print(df['tinggi'].describe())

mean = df['tinggi'].mean()
max_val = df['tinggi'].max()
min_val = df['tinggi'].min()
median = df['tinggi'].median()

print(f"Rata-rata tinggi : {mean:.2f} cm")
print(f"Tertinggi        : {max_val} cm")
print(f"Terendah         : {min_val} cm")
print(f"Median           : {median} cm")
print("")
print("")

print("=== Calon di atas rata-rata tinggi ===")
df_tinggi = df[df['tinggi'] > mean]
print(df_tinggi)
print("")
print("")

print("=== Urutan dari Tertinggi ke Terendah ===")
df_sorted = df.sort_values(by='tinggi', ascending=False)
print(df_sorted)
print("")
print("")

def kategori_tinggi(x):
    if x >= 180:
        return "Tinggi"
    elif x >= 170:
        return "Sedang"
    else:
        return "Pendek"

df['kategori'] = df['tinggi'].apply(kategori_tinggi)

print("=== Data dengan Kategori ===")
print(df)

print("\n=== Jumlah Calon per Kategori ===")
print(df['kategori'].value_counts())
print("")
print("")

print("=== Statistik Lanjutan ===")
std = df['tinggi'].std()
var = df['tinggi'].var()

print(f"Standar Deviasi : {std:.2f}")
print(f"Variansi        : {var:.2f}")
print("")
print("")

print("=== Calon dengan Tinggi Badan Tertinggi ===")
max_height = df['tinggi'].max()
tallest = df[df['tinggi'] == max_height].iloc[0]
print(f"Nama   : {tallest['nama']}")
print(f"Alamat : {tallest['alamat']}")
print(f"Tinggi : {tallest['tinggi']} cm")
```
